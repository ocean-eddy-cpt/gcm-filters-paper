This notebook produces the figure in the "spatially varying filter properties" section of the paper.

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
%config InlineBackend.figure_format = 'retina' # higher res notebook output
```

```python
p = lambda s: (1-0.7*s)*(1-0.8*s)*(1-0.9*s)*(1-s)*(1-1.1*s)*(1-1.2*s)*(1-1.3*s)
k = np.linspace(0,2,1001)
```

```python
#fig, ax = plt.subplots(figsize=[11,7])
plt.plot(k,p(k**2),linewidth=3)
plt.plot(k,p(2*k**2),linewidth=3)
plt.plot(k,p(0.5*k**2),linewidth=3)
plt.xlim([0,2])
plt.xlabel(r'$k$', fontsize=16)
plt.ylim([-1,1])
plt.ylabel(r'$p(\kappa k^2)$', fontsize=16)
plt.grid()
plt.legend([r'$\kappa=1$',r'$\kappa=2$',r'$\kappa=0.5$'], fontsize=16)
plt.tight_layout()
#plt.savefig('Figure2.pdf')
plt.show()
```
